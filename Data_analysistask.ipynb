```python
result = expected_df.merge(company_rates, left_on='Delivery Zone as per X', right_on='Zone', how='left')
expected_df['Weight slab as per X (KG)']=result['Weight Slabs']
```

```python
def calculate(row):
    
    zone = row['Delivery Zone as per X']
    weight_slab_x = row['Weight slab as per X (KG)']
    weight_as_per_x=row['total_weight_as_per_x']
    zone_data = company_rates[company_rates['Zone'] == zone].iloc[0]
    
    weight_slab = zone_data['Weight Slabs']
    fixed_charge = zone_data['Forward Fixed Charge']+zone_data['RTO Fixed Charge']
    additional_charge = zone_data['Forward Additional Weight Slab Charge']+zone_data['RTO Additional Weight Slab Charge']
    
    total_charges = fixed_charge
    
    if weight_as_per_x > weight_slab:
        additional_weight = weight_as_per_x - weight_slab
        additional_charge_amount = (additional_weight / weight_slab) * additional_charge
        total_charges += additional_charge_amount
    
    return total_charges
    
    
```

```python
expected_df['Total Charges as per X'] = expected_df.apply(calculate, axis=1)
expected_df
```
